```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## 1. Load Image

```python
bird=cv2.imread('bird.jpg')
print(bird.shape)
plt.imshow(bird)
```

## 2. Grayscale

```python
bird_gray=cv2.cvtColor(bird,cv2.COLOR_BGR2GRAY)
print(bird_gray.shape)
plt.imshow(bird_gray,cmap='gray')
```

## 3. Resize image

```python
bird_resize1=cv2.resize(bird,(100,100))
bird_resize2=cv2.resize(bird,(100,100),interpolation=cv2.INTER_LINEAR)
bird_resize3=cv2.resize(bird,(100,100),interpolation=cv2.INTER_NEAREST)
bird_resize4=cv2.resize(bird,(100,100),interpolation=cv2.INTER_CUBIC)
print(bird_resize1.shape)
plt.figure(figsize=(15,12))
plt.subplot(141)
plt.title("Normal_resize")
plt.axis('off')
plt.imshow(bird_resize1)
plt.subplot(142)
plt.title("Linear_resize")
plt.axis('off')
plt.imshow(bird_resize2)
plt.subplot(143)
plt.title("Nearest_resize")
plt.axis('off')
plt.imshow(bird_resize3)
plt.subplot(144)
plt.title("Cubic_resize")
plt.axis('off')
plt.imshow(bird_resize4)
```

## 4. Convert color spaces


```python
bird_BGR=cv2.cvtColor(bird,cv2.COLOR_BGR2RGB)
bird_HSV=cv2.cvtColor(bird,cv2.COLOR_RGB2HSV)
bird_LAB=cv2.cvtColor(bird,cv2.COLOR_RGB2LAB)
plt.figure(figsize=(15,12))
plt.subplot(141)
plt.axis('off')
plt.title("Original Image")
plt.imshow(bird)
plt.subplot(142)
plt.axis('off')
plt.title("BGR_Image")
plt.imshow(bird_BGR)
plt.subplot(143)
plt.axis('off')
plt.title("HSV_Image")
plt.imshow(bird_HSV)
plt.subplot(144)
plt.axis('off')
plt.title("LAB_Image")
plt.imshow(bird_LAB)
```

## 6. Rotate Image

```python
bird_rotate_90_clockwise=cv2.rotate(bird,cv2.ROTATE_90_CLOCKWISE)
bird_rotate_180=cv2.rotate(bird,cv2.ROTATE_180)
bird_rotate_90_counterclock=cv2.rotate(bird,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.figure(figsize=(15,12))
plt.subplot(131)
plt.axis('off')
plt.title("90-Clockwise")
plt.imshow(bird_rotate_90_clockwise)
plt.subplot(132)
plt.axis('off')
plt.title("180 rotate")
plt.imshow(bird_rotate_180)
plt.subplot(133)
plt.axis('off')
plt.title("90-counterClockwise")
plt.imshow(bird_rotate_90_counterclock)
```

## 8. Draw Shapes

```python
canvas=bird.copy()
cv2.rectangle(canvas,(10,100),(200,300),(0,0,255),2)
cv2.circle(canvas,(width//2,height//2),100,(255,0,0),2)
cv2.line(canvas,(0,height//2),(width//2,height//2),[255,255,0],2)
cv2.putText(canvas,"This is my bird!!",(300,400),cv2.FONT_HERSHEY_PLAIN,2,(0,255,0),2)
plt.imshow(canvas)
```
